# Linear Regression 


This Notebook tackles the problem of linear regression. 

Linear regression is used to approximate data using a hypothesis $h(x) = \theta_0 + \theta_1x + \dots \theta_n x$. 
The objective is to find the parameters $\theta_0 \dots \theta_n$ that will allow us to predict the output for an unknown new input which is not part of the training data.
Therefore we must define a cost function $J(\theta)$ that captures the error between the real data and our linear function or more general, model. We can numerically compute the needed model parameters $\theta_0 \dots \theta_n$ by running an optimization algorithm such as gradient descent.
Another possible way to determine $\theta_0 \dots \theta_n$ is to use the so-called normal equation.

In this notebook, we will discuss and implement both methods in order to increase the educational benefit.

### 1.4 Cost function:
The data consists of only one feature $X$. Therefore the hypothesis for the linear regression problem has the form 
$h(x) = \theta_0 + \theta_1x$. 
We define the cost function such that it captures the least square error between our prediction and the actual data:

$J(\theta)=\frac{1}{2m} \sum_{i= 1}^{m} (h(x)^{(i)}-y^{(i)})^2.$ 


```python
# Compute Cost
def computeJ(X, y, theta):
    dt = np.dot(X, theta) - y
    return (1/(2*m))*(np.dot(dt, dt))  
```

### 1.5 Gradient desecent algorithm
In order to determine the needed parameters $\theta_0,\theta_1$, we must minimize the partial deriative cost function $\frac{\partial }{\partial \theta_j}J(\theta)$. A commonly used algorithm for minimizing an arbitrary cost function is gradient desecent. It is important to note that the cost function that shall be minimized must be convex, otherwise gradient descent might end up in local minimum.

##### Repeat until convergence:
$\>$$\>$$\>$$\>$$\>$$\>$ $\theta_j := \theta_j-\alpha \frac{\partial }{\partial \theta_j}J(\theta)$

where $ \frac{\partial }{\partial \theta_j}J(\theta) = \frac{1}{m}\sum_{i=1}^{m} ( h_\theta (x^{(i)})-y^{(i)})x^{(i)}_{j} $





### 1.6 Naive Implementation of gradient desecent using loops:

```python
#Gradient Descent Naive Implementation
def gradientDescentNaive(X, y, theta, alpha, iterations):
    
    # Init theta_0, theta_1 = 0
    theta_0 = 0
    theta_1 = 0

    m = len(y)     # number of data samples

    num_iterations = 1500
    J_history = []

    for a in range(num_iterations):
        sum_theta_0 = 0
        sum_theta_1 = 0

        # Compute sum of d/dJ(theta_0), d/dJ(theta_1) 
        for i in range(len(x)):
            sum_theta_0 += (X[i,0]*theta_0 + theta_1*X[i,1]-y[i])*X[i,0]
            sum_theta_1 += (X[i,0]*theta_0 + theta_1*X[i,1]-y[i])*X[i,1]

        # Compute new theta_0, theta_1 
        theta_0 = theta_0 - (alpha/m) * sum_theta_0
        theta_1 = theta_1 - (alpha/m) * sum_theta_1

        # Compute cost for current theta_0, theta_1 and store it in the cost history
        J_history.append(computeJ(X, y, np.array([theta_0,theta_1])))

    return np.array([theta_0,theta_1]), J_history
```

### 1.7 Vectorized implementation of gradient desecent:
The naive implementation has the major drawback that loops are used to iterate through all data samples which is computationally not as efficient as using matrix multiplication.

The Algorithm in it's vectorized form can be implemented as follows:

$\theta := \theta-\alpha \delta $ , with $\delta = $ $\frac{1}{m} \sum_{i= 1}^{m} (h(x)^{(i)}-y^{(i)})x^{(i)}_j.$

The term $\delta = $  $\frac{1}{m} \sum_{i= 1}^{m} (h(x)^{(i)}-y^{(i)})x^{(i)}_j$ can be expressed using vector multiplication as
$X^T(X\theta-y)$ with feature vector $X$, model parameter vector $\theta$ and output vector $y$. Index $j$ indicates the $j$-th feature. 

#### Closed vectorized form of gradient descent:

$\theta := \theta-\alpha \frac{1}{m} X^T(X\theta-y).$

```python
# Vctorized Gradient Descent Implementation
# X[:,1] == 1 by definition  
def gradientDescentVectorized(X, y, theta, alpha, iterations):
    # number of training samples
    m = len(y)

    # cost function history
    J_history = np.zeros(iterations)

    theta_new = theta.copy()

    for i in range(iterations):

        # update theta
        theta_new -= alpha * np.dot(X.T, (np.dot(X, theta_new) - y)) / m        
        
        # store cost 
        J_history[i] = computeJ(X, y, theta_new)

    return theta_new, J_history
```

## 1.8 Linear Regression Experiment 
Initialize feature matrix $X$ and compute cost for initial $\theta_0, \theta_1 = 0$. 

Important: $X_0$ is a column vector of ones


```python
# Init feature vector X 
X = np.ones((m, 2))
X[:,1] = data[:,0] 

# init model parameters with theta = [0,0]
theta = np.zeros(2)

# compute the cost function for theta = [0,0]
print("Initial Cost (squarred error) using theta_0,theta_1 = 0:  ",computeJ(X, y, theta))
```

### 1.8. 1 Perform Naive Gradient Descent:

```python
# Init learning rate alpha and number of iterations
alpha = 0.01
num_iterations = 1500

# run gradient descent
theta, J_history = gradientDescentNaive(X, y, theta, alpha, num_iterations)
print("Theta: ", theta)
```

### 1.8.2 Perform Vectorized Gradient Descent:

```python
# Run gradient descent
theta=np.zeros(2)
J_history = []
theta, J_history = gradientDescentVectorized(X, y, theta, alpha, num_iterations)
print("Theta: ", theta)
```

### Conclusion 1.8.1 vs 1.8.2
The results show that both methods provide the same model parameters $\theta$

## 2 Linear regression with multiple features

```python
data = np.loadtxt('ex1data2.txt', delimiter=',')
x, y = data[:,:-1], data[:,-1]
x_feature_1 = x[:,0]
x_feature_2 = x[:,1]
m = len(y)
```

### 2.1 Show first 10 rows  to get a better feeling for the data:

### 2.2 Plot multivariant data (Two features):

### Data Description
The data table as well as the plots of the two feature vectors $X_1$ and $X_2$ show that the values of $X_1$ are approximately 10 times bigger than the values of $X_2$. In this case the contour lines will be stretched vertically and clinched horizontally which means that gradient descent will run extremely slow. This is why in practice feature scaling becomes very important. The basic idea is to transform the values into a similiar range.

Further Information:
https://www.quora.com/In-gradient-descent-without-feature-scaling-why-does-theta-descend-quickly-for-small-ranges-and-slowly-for-large-ones



### 2.3 Feature Scaling
A commonly used technique to perform feature scaling is called mean normalization. In order to scale the features 

```python
def featureNormalize(X):
    
    # compute mean and standard deviation
    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    
    X_norm = (X - mu) / sigma
    
    return X_norm, mu, sigma
```

```python
# normalize features
X, mu, sigma = featureNormalize(x)

# create feature vector that consists of zero feature column 
# of ones and the (now normalized) feature matrix X 
X = np.hstack((np.ones((X.shape[0],1)), X))
```

## 2.4 Gradient Descent Experiment  

```python
# gradient descent values
alpha = 0.01
iterations = 2500
theta = np.zeros(3)

# run gradient descent
theta, J_history = gradientDescentVectorized(X, y, theta, alpha, iterations)
print(theta)
```

### 2.4.1 Plot cost function $J(\theta$):

```python
#Plot Cost Function
print("Theta after optimization:   ", theta)
print("Cost (squarred error) new theta0,theta1,theta2:  ",computeJ(X, y, theta))
plt.plot(1)
plt.xlabel("Iteration")
plt.ylabel(r'J($\theta_0, \theta_1, \theta_2$)')
plt.xlim([0,len(J_history)])
print("Cost Max:  ", max(J_history))
print("Cost Min:  ", min(J_history))

j_hist, = plt.plot(J_history, color="green", label= "Cost Function")
plt.legend([j_hist])
plt.legend()
plt.grid()
```

### 2.4.2 Prediction:


```python
# Estimate the price of a 1650 sq-ft, 3 br house
vals = (np.array([1650, 3]) - mu) / sigma
vals = np.hstack((np.array([1]), vals))

predict = np.dot(theta, vals)
print ("A 1650 sq-ft 3 br house will cost $", predict)
```

### 2.4.2 Plot data and prediction:


```python
figure, ax = plt.subplots(1, 2, figsize=(14,4))

ax[0].plot(x[:,0], y, 'o', color='g')
ax[0].plot(1650, predict, 'x', color='r', mew=2)
ax[0].set_xlabel(r'Size of house [sq ft]')
ax[0].set_ylabel(r'Price of house')
ax[0].grid()

ax[1].plot(x[:,1], y, 'o', color='y')
ax[1].plot(3, predict, 'x', color='r', mew=2)
ax[1].set_xlabel(r'Number of bedrooms')
ax[1].set_ylabel(r'Price of house')
ax[1].grid()
```

### Learning Rate $\alpha$:
The learning rate $\alpha$ describes the step size that gradient descent takes towards the global minimum. While a small learning rate results in a high number of iterations, a too big $\alpha$ can result in divergence. The right choice of $\alpha$ is not trivial and in practice the algorithm needs to be performed  a few times to pick a apropriate learning rate.

```python

alphas = [0.9, 0.3, 0.1, 0.03, 0.01]
iterations = 100
for alpha in alphas:
    theta_new, J_history = gradientDescentVectorized(X, y, theta, alpha, iterations)
    plt.plot(J_history, label='learning rate={0}'.format(alpha))

plt.ylabel(r"$J(\theta)$")
plt.xlabel(r"Number of Iterations")
plt.legend(loc="best")
plt.grid()
```

## 3. Normal Equation:
When the number of features is small (100s-10000s) we can use the Normal Equation to determine our model parameters $\theta$. Using the normal equation there is no need to choose a learning rate which is a big advantage.
However, if the number of features is big (> Mio features) the normal equation becomes computationally expensive.
As discussed earlier, feature scaling makes the optimization better conditioned and leads to faster and better convergence.
So in theory, when you are using the normal equation, you don’t use numerical methods, like gradient descent. Therefore, feature scaling is not necessary when using the normal equation. 


#### Normales Equation

$\theta := (X^TX)^{-1} (X^Ty)$



```python
def normalEquation(X, y):
    
    inv_term = np.linalg.inv(np.dot(X.T, X))
    theta = np.dot(inv_term, np.dot(X.T, y))
    
    return theta
```

### 3.1 Prepare data (no feature scaling needed when using the normal equation)

```python
# re-compute X with out normalization
X = np.hstack((np.ones((x.shape[0],1)), x))

# get solution with normal equation
theta_norm_eq = normalEquation(X, y)
```

### 3.2 Prediction


```python
# Estimate the price of a 1650 sq-ft, 3 br house
vals = np.array([1650, 3])
vals = np.hstack((np.array([1]), vals))

# this is the same answer as above if run for more iterations
predict = np.dot(theta_norm_eq, vals)
print ("A 1650 sq-ft 3 bedroom house will cost $%6.2f", predict)
```
