# Linear Regression #


This Notebook takles the problem of linear regression. Linear regression is used to approximate data using a linear function $h(x) = \theta_0 + \theta_1x$. Therefore the objective is to find the parameters $\theta_0,\theta_1$.

### Import needed modules:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

### Import the dataset: 

```python
data = np.loadtxt('ex1data1.txt', delimiter=',')
x, y = data[:,0], data[:,1]
m = len(y)
```

### Plot data:

```python
fig0 = plt.figure(0)
dataset, = plt.plot(x,y,'o', color='r', label="Training Data")
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.legend([dataset])
plt.legend()
plt.grid()
```

### Define a cost function:
The data consists of only one feature $X$. Therefore the hypothesis for the linear regression problem has the form 
$h(x) = \theta_0 + \theta_1x$. 
To minimize the least square error between our prediction and the actual data, we use the cost function $J(\theta)=\frac{1}{2m} \sum_{i= 1}^{m} (h(x)^{(i)}-y^{(i)})^2.$ 


```python
# Compute Cost
def computeJ(X, y, theta):
    dt = np.dot(X, theta) - y
    return (1/(2*m))*(np.dot(dt, dt))  
```

### Gradient desecent
In order to determine the needed parameters $\theta_0,\theta_1$, we mustminimize the cost function $J(\theta)$. A commonly used algorithm for minimizing an arbitrary cost function is gradient desecent:

##### Repeat until convergence:
$\>$$\>$$\>$$\>$$\>$$\>$ $\theta_j := \theta_j-\alpha \frac{\partial }{\partial \theta_j}J(\theta)$




### Naive Implementation of gradient desecent using loops:

```python
#Gradient Descent Naive Implementation
def gradientDescentNaive(X, y, theta, alpha, iterations):
    
    # Init theta_0, theta_1 = 0
    theta_0 = 0
    theta_1 = 0
    
    m = len(y)     # number of data samples
    
    print("XSHAPE ", X.shape)

    num_iterations = 1500
    J_history = []
    
    for a in range(num_iterations):
       sum_theta_0 = 0
       sum_theta_1 = 0
       
       # Compute sum of d/dJ(theta_0), d/dJ(theta_1) 
       for i in range(len(x)):
         sum_theta_0 += (X[i,0]*theta_0 + theta_1*X[i,1]-y[i])*X[i,0]
         sum_theta_1 += (X[i,0]*theta_0 + theta_1*X[i,1]-y[i])*X[i,1]
       
       # Compute new theta_0, theta_1 
       theta_0 = theta_0 - (alpha/m) * sum_theta_0
       theta_1 = theta_1 - (alpha/m) * sum_theta_1
       
       # Compute cost for current theta_0, theta_1 and store it in the cost history
       J_history.append(computeJ(X, y, np.array([theta_0,theta_1])))

    return np.array([theta_0,theta_1]), J_history
```

### Vectorized implementation of gradient desecent:
The naive implementation has the major drawback that loops are used to iterate through all data samples which is computationall not as efficient as using matrix multiplication.

The Algorithm in it's vectorized form can be implemented as follows:

$\theta := \theta-\alpha \delta $ , with $\delta = $ $\frac{1}{m} \sum_{i= 1}^{m} (h(x)^{(i)}-y^{(i)})x^{(i)}_j.$

The term $\delta = $  $\frac{1}{m} \sum_{i= 1}^{m} (h(x)^{(i)}-y^{(i)})x^{(i)}_j$ can be expressed using vector multiplication as
$X^T(X\theta-y)$ with feature vector $X$, model parameter vector $\theta$ and output vector $y$. Index $j$ indicates the $j$-th feature. However, as our dataset has only one feature $j$ is equal to 1. 

```python
# Vctorized Gradient Descent Implementation
# X[:,1] == 1 by definition  
def gradientDescentVectorized(X, y, theta, alpha, iterations):
    # number of training samples
    m = len(y)

    # cost function history
    J_history = np.zeros(iterations)

    theta_new = theta.copy()

    for i in range(iterations):

        # update theta
        theta_new -= alpha * np.dot(X.T, (np.dot(X, theta_new) - y)) / m        
        
        # store cost 
        J_history[i] = computeJ(X, y, theta_new)

    return theta_new, J_history
```

## Linear Regression Experiment

### Initialize feature vector $X$ and compute cost for initial $\theta_0, \theta_1 = 0$:

```python
# Iniit feature vector X 
X = np.ones((m, 2))
X[:,1] = data[:,0] 

# init model parameters with theta = 0
theta = np.zeros(2)

# compute the cost function for theta = 0
print("Initial Cost (squarred error) using theta_0,theta_1 = 0:  ",computeJ(X, y, theta))
```

### Perform Naive Gradient Descent:

```python
# Init learning rate alpha and number of iterations
alpha = 0.01
num_iterations = 1500

# run gradient descent
theta, J_history = gradientDescentNaive(X, y, theta, alpha, num_iterations)
print("Theta: ", theta)
```

### Perform Vectorized Gradient Descent;

```python
# Run gradient descent
theta=np.zeros(2)
J_history = []
theta, J_history = gradientDescentVectorized(X, y, theta, alpha, num_iterations)
print("Theta: ", theta)
```

### Plot cost function $J(\theta$):

```python
#Plot Cost Function
print("Theta after optimization:   ", theta)
print("Cost (squarred error) new theta0,theta1:  ",computeJ(X, y, theta))
plt.plot(1)
plt.xlabel("Iteration")
plt.ylabel(r'J($\theta_0, \theta_1$)')
plt.xlim([0,len(J_history)])
print("Cost Max:  ", max(J_history))
print("Cost Min:  ", min(J_history))

j_hist, = plt.plot(J_history, color="green", label= "Cost Function")
plt.legend([j_hist])
plt.legend()
plt.grid()
```

### Visualize regression line:

```python
# Plot Libe
print("Theta   ", theta)
plt.plot(2)
dataset, = plt.plot(X[:,1], y, 'o', color='r', label="Training Data")
line, = plt.plot(X[:,1], theta[0]+theta[1]*X[:,1],  color='b', label="Regression Line" )

plt.xlim([min(X[:,1]),max(X[:,1])])
plt.xlabel('Population of City in 10,000s')
plt.ylabel('Profit in $10,000s')
plt.legend([dataset, line])
plt.legend()
plt.grid()
plt.show()
```

```python
### Plot meshgrid
```

```python

# Prepare grid for cost visualization
theta0_vals = np.linspace(-5, 5, 100)
theta1_vals = np.linspace(-5, 5, 100)

# Init Cost values = 0
J_vals = np.zeros((len(theta0_vals), len(theta1_vals)))

# Compute cost values for grid
for i, t0 in enumerate(theta0_vals):
    for j, t1 in enumerate(theta1_vals):
        t = np.array([t0, t1])
        J_vals[i,j] = computeJ(X, y, t)

# Create surface plot
fig = plt.figure(3)
ax = fig.gca(projection='3d')
T0, T1 = np.meshgrid(theta0_vals, theta1_vals)
surf = ax.plot_surface(T0, T1, J_vals.T, cmap='jet', rstride=3,
                       cstride=3, antialiased=True)
ax.set_xlabel(r'$\theta_0$')
ax.set_ylabel(r'$\theta_1$')
plt.show()

```

```python

plt.plot(4)
# make contour plot
plt.contour(T0, T1, J_vals.T, np.logspace(-2, 3, 20))
plt.plot(theta[0], theta[1], 'x', color='r', mew=2)
plt.xlabel(r'$\theta_0$')
plt.ylabel(r'$\theta_1$')
plt.grid()

plt.show()
```
